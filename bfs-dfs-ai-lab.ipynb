# BFS Implementation

```python
from queue import Queue

def bfs(graph, start_node):
    visited = set()
    q = Queue()

    q.put(start_node)
    visited.add(start_node)

    while not q.empty():
        node = q.get()
        print(node, end=" ")

        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                visited.add(neighbor)
                q.put(neighbor)


graph = {
    'A': ['B','C','D'],
    'B': ['A','E','F'],
    'C': ['A','G'],
    'D': ['A','H'],
    'E': ['B'],
    'F': ['B'],
    'G': ['C'],
    'H': ['D']
}

print("BFS Traversal Order:")
bfs(graph, 'A')
```

# DFS Implementation

```python
def dfs(graph, start_node):
    visited = set()
    stack = [start_node]

    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)

            for neighbor in reversed(graph.get(node, [])):
                if neighbor not in visited:
                    stack.append(neighbor)


graph = {
    'A': ['B','C','D'],
    'B': ['A','E','F'],
    'C': ['A','G'],
    'D': ['A','H'],
    'E': ['B'],
    'F': ['B'],
    'G': ['C'],
    'H': ['D']
}


print("DFS Traversal Order: ")
dfs(graph, 'A')
```
