# **Breadth First Search**

In this block the BFS is implemented in a normal way just by importing dequeue from the prthon library

```python
#importing the library
from collections import deque


def bfs(graph,start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.pop()
        if node not in visited:
            visited.add(node)
            print(node, end = " ")
            queue.extend(graph[node] - visited)
            
#creatibg the graph
graph = {
    'A': {'B', 'C','D'},
    'B': {'E','F'},
    'C': {'G'},
    'D': {'H'},
    'E': {'B' },
    'F': { 'B'},
    'G': { 'C'},
    'H': {'D' }
}
bfs(graph,'A')
```

In this block the BFS is implemented by using networkx library

```python
import networkx as nx

# Create a graph
G = nx.Graph()
edges = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('E', 'F')]
G.add_edges_from(edges)

# Perform BFS starting from node 'A'
bfs_result = list(nx.bfs_edges(G, source='A'))
bfs_nodes = ['A'] + [v for u, v in bfs_result]  # Get nodes in BFS order

print("BFS Traversal:", bfs_nodes)
```

```python
#shortest path finding by using bfs
from collections import deque

def shortest_path(graph,start,target):
    queue = deque([start])
    visited = set()
    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == target:
            return path
        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                new_path = path + neighbor
                queue.append(new_path)
    
                
#creatibg the graph
graph = {
    'A': {'B', 'C','D'},
    'B': {'E','F'},
    'C': {'G'},
    'D': {'H'},
    'E': {'B' },
    'F': { 'B'},
    'G': { 'C'},
    'H': {'D' }
}
road = shortest_path(graph,'A','F')
print(road)
```
